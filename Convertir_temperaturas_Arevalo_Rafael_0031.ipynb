## 1. Fórmulas de Conversión

Para entender cómo pasar de una escala a otra, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:** Multiplicamos los grados Celsius por $1.8$ (o $9/5$) y sumamos $32$.
    $$°F = (1.8 \times °C) + 32$$

* **De Fahrenheit a Celsius:** Restamos $32$ a los grados Fahrenheit y luego dividimos el resultado entre $1.8$.
    $$°C = \frac{°F - 32}{1.8}$$

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a Fahrenheit (Temperatura ambiente)
1.  **Identificar datos:** $°C = 25$.
2.  **Sustituir en la fórmula:** $°F = (1.8 \times 25) + 32$.
3.  **Multiplicación:** $1.8 \times 25 = 45$.
4.  **Suma final:** $45 + 32 = 77$.
**Resultado:** $25°C$ equivale a **77°F**.

### Ejemplo B: Convertir 98.6°F a Celsius (Temperatura corporal)
1.  **Identificar datos:** $°F = 98.6$.
2.  **Sustituir en la fórmula:** $°C = (98.6 - 32) / 1.8$.
3.  **Resta:** $98.6 - 32 = 66.6$.
4.  **División final:** $66.6 / 1.8 = 37$.
**Resultado:** $98.6°F$ equivale a **37°C**.

---

```python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== MENÚ DE CONVERSIONES ===")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Arevalo Rafael NL = 3")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada a la función principal para iniciar el programa
menu()
```
