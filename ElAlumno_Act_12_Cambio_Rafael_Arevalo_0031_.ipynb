### Las variables
1. $CP = Costo del producto$

2. $CD = Cantidad de dinero$
3. $Cambio --> Dinero restante (resultado)$

### Las formulas
$Cambio = CD - CP$

$CP > CD$

```python
# Determinando el mayor usando if else
# Datos de entrada A , B
# Mostrar el mayor si es falso
# si es verdadero asignar A a M

A = float(input("Ingrese el valor A: "))
B = float(input("Ingrese el valor B: "))

if A > B:
    M = A
else:
    M = B

print("El valor mayor es: ", M)

print("Fin del programa")
print("Programa realizado por Arevalo Rafael NL = 3")
```
